## Feature engineering

A feature is a measurable property in a dataset, and a feature can be used as an input to a machine learning mode
- if we want a model that predicts precipitation at a given place and time, we might use temperature, humidity, month, altitude, etc. as inputs. These are our features.

Feature engineering helps us make models better
- performnace (how accurate)
- runtime (how long it takes to run)
- intepretability (gives us insights)
- generalizability (works with new data)



Three types of feeture engineering techniques

**Feature transformation methods**
- applied before implementing amodel
- eg. scaling, binning, logarythimic transformations, hashing, one hot encoding
- improve performance, runtime and interpretability of a model.

**Dimensionality reduction methods**
- Dimensionality of a dataset refers to the number of features within a dataset
- eg.a hundred-feature problem can be reduced to less than ten modified features, saving a lot of computational time and resources while maintaining or even improving performance
- some are ML models themselves, such as Principal Component Analysis (PCA), Linear Discriminant Analysis (LDA)
- These techniques transform the existing feature space into a new subset of features that are ordered by decreasing importance. Since they “extract” new features from high dimensional data they’re also referred to as Feature Extraction methods. The transformed features do not directly relate to anything in the real world anymore. Rather, they are mathematical objects that are related to the original features. However, these mathematical objects are often difficult to interpret. The lack of interpretability is one of the drawbacks of dimensionality reduction.

**Feature Selection methods**
-  allow us to choose among the pool of features available
- A.Filter methods: filter out useful feautures. eg. correlation coefficients (Pearson, Spearman, etc) , chi^2, ANOVA, and Mutual Information calculations.
- B.Wrapper methods: “greedy search strategy”. They pick a subset of features, train a model, evaluate it, try a different subset of features, train a model again, and so on until the best possible set of features and most optimal performance is reached. eg. Forward Feature Selection, Backward Feature Elimination and Sequential Floating
- C.Embedded methods: implemented during the model implementation step. Regularization techniques such as Lasso or Ridge tweak the model to get it to generalize better to new and unknown data. Tree-based feature importance is another widely used embedded method. This method provides insight into the features that are most relevant to the outcome variable while fitting decision trees to the data.
