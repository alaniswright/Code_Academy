```python
# Codeacademy portfolio project for data visualisation
# https://www.codecademy.com/paths/data-science-nlp/tracks/dsf-data-visualization-portfolio-project/modules/dsf-data-visualization-portfolio-project

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("all_data.csv")
display(df.describe())
display(df.info())
display(df.shape) # rows and columns
display(df.columns) # column names
display(df.dtypes) # 
```

```python
my_plot = sns.boxplot(x="Country", y="Life expectancy at birth (years)", data=df)
my_plot.set_xticklabels(my_plot.get_xticklabels(), rotation=45)
my_plot.set(title='Life expectancy variation by Country')
plt.show()
plt.clf()
```

```python
my_plot = sns.lineplot(data=df, x = "Year", y = "GDP", hue="Country")
my_plot.set(title='GDP change by year by Country')
plt.xticks(rotation=45)
plt.show()
plt.clf()
```

```python
my_plot = sns.lineplot(data=df, x = "Year", y = "Life expectancy at birth (years)", hue="Country")
my_plot.set(title='Life expectancy by year by Country')
plt.xticks(rotation=45)
plt.show()
plt.clf()

countries_unique = list(set(df["Country"]))
countries_unique.sort()

for c in countries_unique:
    print(c)
```
